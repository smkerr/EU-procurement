## Graph-tool

How to visualize using graph-tool.

```python
import graph_tool.all as gt
import matplotlib.pyplot as plt
```

```python
g_gt = gt.collection.ns["urban_streets/ahmedabad"]
```

```python
gt.graph_draw(g_gt, output_size=(500, 500), vertex_color=[1,1,1,0],
           vertex_size=1, edge_pen_width=1.2,
           vcmap=matplotlib.cm.gist_heat_r)
```

```python
hist = gt.vertex_hist(g_gt, "out")

plt.plot(hist[1][:-1], hist[0])

plt.show()
```

```python
g_gt = gt.GraphView(g_gt, vfilt=gt.label_largest_component(g_gt))

pr = gt.pagerank(g_gt)

gt.graph_draw(g_gt, vertex_fill_color=pr,

              vertex_size=gt.prop_to_size(pr, mi=1, ma=15),

              vorder=pr, vcmap=matplotlib.cm.gist_heat)
```

```python
g = gt.collection.ns["copenhagen/sms"]
```

```python
g.save("copenhagen_sms.graphml")
```

```python
gt.graph_draw(g, output_size=(500, 500), vertex_color=[1,1,1,0],
           vertex_size=1, edge_pen_width=1.2,
           vcmap=matplotlib.cm.gist_heat_r)
```

```python
hist = gt.vertex_hist(g, "out")

plt.plot(hist[1][:-1], hist[0])

plt.show()
```

```python
state = gt.minimize_blockmodel_dl(g)
state.draw()
```
